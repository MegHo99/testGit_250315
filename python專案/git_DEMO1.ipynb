```python
import cmath

a = float(input('輸入 a: '))
b = float(input('輸入 b: '))
c = float(input('輸入 c: '))

# 計算判別式
d = (b**2) - 4*(a*c)

# 根據判別式的值找出根
if d > 0:
    sol1 = (-b - cmath.sqrt(d))/(2*a)
    sol2 = (-b + cmath.sqrt(d))/(2*a)
    print('這個方程式有兩個不同的實數根： {} 和 {}'.format(sol1,sol2))
elif d == 0:
    sol1 = (-b - cmath.sqrt(d))/(2*a)
    print('這個方程式有一個重複的實數根： {}'.format(sol1))
else:
    sol1 = (-b - cmath.sqrt(d))/(2*a)
    sol2 = (-b + cmath.sqrt(d))/(2*a)
    print('這個方程式有兩個複數根： {} 和 {}'.format(sol1,sol2))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍和間隔
x = np.linspace(0, 2*np.pi, 100)

# 計算 y 值 (sin(x))
y = np.sin(x)

# 畫出正弦波
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('Sin plot')
plt.xlabel('x')
plt.ylabel('sin(x)')

# 顯示圖表
plt.show()
```

### cos plot

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍和間隔
x = np.linspace(0, 2*np.pi, 100)

# 計算 y 值 (cos(x))
y = np.cos(x)

# 畫出餘弦波
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('Cosine plot')
plt.xlabel('x')
plt.ylabel('cos(x)')

# 顯示圖表
plt.show()
```
